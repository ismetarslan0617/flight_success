```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('csv_data/test_corrupted.csv',encoding = 'utf-8')
```

```python
df
```

```python
print(df.isna().sum())
```

```python
print("\nRüzgar hızı ortalaması",df['RÜZGAR HIZI (KM/SAAT)'].mean())

print("\nSıcaklık (°C) ortalaması",df['Sıcaklık (°C)'].mean())

print("\nNem Seviyesi (%) ortalaması",df['Nem Seviyesi (%)'].mean())

print("\nYağmur Olasılığı (%)",df['Yağmur Olasılığı (%)'].mean())
```

```python
doluluk_orani = df.notna().mean() * 100
doluluk_orani
```

```python
plt.figure(figsize=(10, 6))

ax = doluluk_orani.plot(kind='bar', color='#156082', edgecolor='black')

plt.title('Her Sütunun Doluluk Oranı (%)', fontsize=16, fontweight='bold')
plt.ylabel('Doluluk Orani (%)', fontsize=12)
plt.xlabel('Sütunlar', fontsize=12)

plt.xticks(rotation=0, fontsize=11)
plt.yticks(fontsize=11)

plt.grid(axis='y', linestyle='--', alpha=0.7)

for i in ax.containers:
    ax.bar_label(i, label_type='edge', padding=3, fontsize=12)

plt.ylim(0, doluluk_orani.max() * 1.1)
plt.tight_layout()
plt.show()
```
