```python
import pandas as pd
import numpy as np

# Veri setini yükleme
df = pd.read_csv("test_IQR.csv")

# Bozulacak sütunların listesi
columns_to_corrupt = ["RÜZGAR HIZI (KM/SAAT)", "Yağmur Olasılığı (%)", "Nem Seviyesi (%)","Sıcaklık (°C)"]

# Veri setindeki satır sayısını al
num_rows = len(df)

# Rastgele bozulacak 10 satır seç
rows_to_corrupt = np.random.choice(num_rows, size=40, replace=False)

# Her seçilen satır için rastgele bir sütunu boz
for row in rows_to_corrupt:
    col_to_corrupt = np.random.choice(columns_to_corrupt)
    df.loc[row, col_to_corrupt] = np.nan

# Değişiklikleri kontrol etme
print(df.loc[rows_to_corrupt])

# Bozulmuş veri setini yeni bir dosyaya kaydetme
df.to_csv("test_corrupted.csv", index=False)

df1 = pd.read_csv('test_corrupted.csv',encoding = 'utf-8')
```
