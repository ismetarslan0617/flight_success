```python
import pandas as pd
import numpy as np

df = pd.read_csv('csv_data/365_days_weather_data.csv',encoding = 'utf-8')
df = df.iloc[:139]

# Assuming `data` is your prediction dataset
df.rename(columns={"Sıcaklık (C°)": "Sıcaklık (°C)"}, inplace=True)

# Başarı oranını çevresel faktörlere göre hesaplayan fonksiyon
def calculate_success_probability(row):
    success_probability = 100  # Başlangıçta başarı oranı 100%

    
    if row['RÜZGAR HIZI (KM/SAAT)'] > 40:
        success_probability -= (row['RÜZGAR HIZI (KM/SAAT)'] - 40) * 2 

    elif row['RÜZGAR HIZI (KM/SAAT)'] > 30:
        success_probability -= (row['RÜZGAR HIZI (KM/SAAT)'] - 30) * 1.5 

    elif row['RÜZGAR HIZI (KM/SAAT)'] > 20:
        success_probability -= (row['RÜZGAR HIZI (KM/SAAT)'] - 20) * 1  

    elif row['RÜZGAR HIZI (KM/SAAT)'] > 10:
        success_probability -= (row['RÜZGAR HIZI (KM/SAAT)'] - 10) * 0.5  

    if row['Yağmur Olasılığı (%)'] > 50:
        success_probability -= (row['Yağmur Olasılığı (%)'] - 50 ) * (4/5) 
    elif row['Yağmur Olasılığı (%)'] > 10:
        success_probability -= (row['Yağmur Olasılığı (%)'] - 10 ) * (5/8)
    else:
        success_probability -= (row['Yağmur Olasılığı (%)'])  

    if row['Nem Seviyesi (%)'] > 60:
        success_probability -= ((row['Nem Seviyesi (%)'] - 60)) * (1/4)  
    elif row['Nem Seviyesi (%)'] < 30:
        success_probability -= ((row['Nem Seviyesi (%)'] )) * (1/3)  
    
    if row['Sıcaklık (°C)'] < 2:
        success_probability -= 10 
    if row['Sıcaklık (°C)'] > 37:
        success_probability -= 10 

    # Başarı oranını 0 ile 100 arasında sınırlama
    success_probability = np.clip(success_probability, 0, 100)

    return success_probability

# Yeni uçuş başarı oranını hesapla ve yeni bir sütun olarak ekle
df['Uçuş Başarısı (%)'] = df.apply(calculate_success_probability, axis=1)

df.to_csv('csv_data/365_days_weather_data_classification.csv', index=False)

# Yeni DataFrame'i yazdırma
print(df)
```
